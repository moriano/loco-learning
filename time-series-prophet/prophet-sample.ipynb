# A simple example using facebook prophet

### Prophet is a machine learning library to perform predictions on a time series data

```python
%matplotlib inline

import pandas as pd
import numpy as np
from fbprophet import Prophet
```

```python
df = pd.read_csv('example_wp_peyton_manning.csv')
df['y'] = np.log(df['y'])
df.head()
```

```python
df.tail()
```

```python
m = Prophet()
m.fit(df);
```

```python
# Python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast);
```

```python
m.plot_components(forecast);
```
